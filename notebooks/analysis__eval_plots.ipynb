```python
from utils import LIMITED_EVAL_METRICS, MODEL_SIZES, get_checkpoint_evals, CHECKPOINT_STEPS 

from matplotlib import pyplot as plt
```

```python
# For plotting out the evaluation results 

for model_size in MODEL_SIZES:
    checkpoint_evals = get_checkpoint_evals(model_size)
    plt.figure()
    for eval_metric in LIMITED_EVAL_METRICS:
        plt.plot(CHECKPOINT_STEPS, checkpoint_evals[eval_metric], label=eval_metric)
    plt.legend()
    plt.xlabel('Step')
    plt.ylabel('Accuracy')
    plt.title(f'{model_size} Zero-shot evaluation')
```
